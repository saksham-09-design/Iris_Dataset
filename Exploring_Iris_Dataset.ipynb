# Dividing data To analyse

```python
data_setosa = data[data['Species']=='Iris-setosa']
data_versicolor = data[data['Species']=='Iris-versicolor']
data_virginica = data[data['Species']=='Iris-virginica']
```

# Calculating necessary values to visualize the data

```python
species = ['Setosa', 'Versicolor', 'Virginica']
```

## Average / Maximum / Minimum sepal Length

```python
s_sep_len_avg = data_setosa['SepalLengthCm'].mean()
ve_sep_len_avg = data_versicolor['SepalLengthCm'].mean()
vi_sep_len_avg = data_virginica['SepalLengthCm'].mean()
avg_list = np.array([s_sep_len_avg, ve_sep_len_avg, vi_sep_len_avg])

s_sep_len_max = data_setosa['SepalLengthCm'].max()
ve_sep_len_max = data_versicolor['SepalLengthCm'].max()
vi_sep_len_max = data_virginica['SepalLengthCm'].max()
max_list = np.array([s_sep_len_max, ve_sep_len_max, vi_sep_len_max])

s_sep_len_min = data_setosa['SepalLengthCm'].min()
ve_sep_len_min = data_versicolor['SepalLengthCm'].min()
vi_sep_len_min = data_virginica['SepalLengthCm'].min()
min_list = np.array([s_sep_len_min, ve_sep_len_min, vi_sep_len_min])
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

titles = ['Average Sepal Length', 'Minimum Sepal Length', 'Maximum Sepal Length']
data_list = [avg_list, min_list, max_list]
colors = ['skyblue', 'lightgreen', 'salmon']

for i in range(3):
    bars = axes[i].bar(species, data_list[i], color=colors[i])
    axes[i].set_title(titles[i])
    axes[i].set_ylabel('Sepal Length (cm)')
    axes[i].grid(axis='y', linestyle='--', alpha=0.6)
    
    # Optional
    for bar in bars:
        axes[i].text(bar.get_x() + bar.get_width()/2, bar.get_height(),
                     f'{bar.get_height():.2f}', 
                     ha='center', 
                     va='bottom', 
                     fontsize=9, 
                     fontweight='bold'
                    )

fig.suptitle('Sepal Length Statistics by Species', fontsize=15, fontweight='bold')
plt.tight_layout()
```

## Let's train the model for pridiction Now

```python
from sklearn import tree
```

```python
iris_data = pd.read_excel('iris-train.xlsx')
iris_data.head()
```

```python
training_data = iris_data.iloc[:,:-1]
training_data.head()
```

```python
training_label = iris_data.iloc[:,-1]
training_label.head()
```

```python
model_Decision = tree.DecisionTreeClassifier()
model_Decision = model_Decision.fit(training_data, training_label)
```

```python
iris_test = pd.read_excel('iris-test.xlsx')
iris_test.head()
```

```python
test_data = iris_test.iloc[:,:-1]
test_data.head()
```

```python
results = model_Decision.predict(test_data)
```

```python
test_labels = iris_test.iloc[:,-1]
test_labels.head()
```

```python
for i in range(len(results)):
    if results[i] == test_labels.iloc[i]:
        print("True", end=" ")
    else:
        print("False", end=" ")
```

```python
from sklearn.ensemble import RandomForestClassifier
model_Random = RandomForestClassifier()
model_Random = model_Random.fit(training_data, training_label)
```

```python
results_random = model_Random.predict(test_data)
```

```python
for i in range(len(results)):
    if results_random[i] == test_labels.iloc[i]:
        print("True", end=" ")
    else:
        print("False", end=" ")
        print(f"\nPridicted: {results_random[i]}")
        print(f"Original: {test_labels[i]}")
```
